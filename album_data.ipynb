```python
import pandas as pd
```

```python
#Read albums' data from tsv

pd_album = pd.read_csv('album_data.tsv', sep='\t')
pd_album
```

```python
pd_album.info()
```

```python
num_rows = len(pd_album)
num_cols = len(pd_album.columns)
num_rows, num_cols
```

```python
dtypes = pd_album.dtypes
dtypes
```

```python
#Count how many duplicated albums' id

pd_album['id'].duplicated().value_counts()
```

```python
#Show off

num_duplicated_rows = pd_album.duplicated()

pd_album[num_duplicated_rows == False]
```

```python
#Count how much time an album occur

dups_album = pd_album.pivot_table(columns=pd_album[['id', 'name']], aggfunc='size')
dups_album
```

```python
#Write albums' occurrences to tsv

new_column_names = ['occurrences']

dups_album.to_csv('album_occurrences.tsv', header=new_column_names, sep='\t')
```

```python
df = pd.read_csv('album_occurrences.tsv', sep='\t')
df
```

```python
#Sort by occurrences

result = df.sort_values(by='occurrences', ascending=False)
result
```

```python
#Drop duplicated rows

pd_album = pd_album.drop_duplicates(subset=['id'], keep='first', inplace=False)
```
