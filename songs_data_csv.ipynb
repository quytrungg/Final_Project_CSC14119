```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
raw_data = pd.read_csv('songs_data.tsv', sep='\t')
raw_data.head()
```

```python
data_np = np.array(raw_data)

for i in range(len(data_np)):
    data_np[i, 2] = data_np[i, 2].split(',')
    data_np[i, 3] = data_np[i, 3].split(',') if type(data_np[i, 3]) != float else ['Untitled']
    data_np[i, 4] = list(map(int, data_np[i, 4].split(',')))
    data_np[i, 5] = list(map(int, data_np[i, 5].split(',')))
    data_np[i, 6] = int(data_np[i, 6])
    data_np[i, 8] = datetime.strptime(data_np[i, 8], '%Y-%m-%d') if '-' in data_np[i, 8] else datetime.strptime(data_np[i, 8], '%Y')
    data_np[i, 9] = int(data_np[i, 9])
    data_np[i, 10] = int(data_np[i, 10])
    data_np[i, 11] = 0 if data_np[i, 11] == False else 1
    data_np[i, 12] = int(data_np[i, 12])
```

```python
songs_data = pd.DataFrame(data_np, columns=raw_data.columns)
songs_data.head()
```

```python
data_np_copy = np.copy(data_np)

for i in range(len(data_np_copy)):
    data_np_copy[i, 4] = int(np.sqrt(np.sum(list(map(lambda x: x**2, data_np_copy[i, 4])))))
    data_np_copy[i, 5] = min(100, int(np.sqrt(np.sum(list(map(lambda x: x**2, data_np_copy[i, 5]))))))

standard_data = np.array([[col for col in line if type(col) == int] for line in data_np_copy])
```

```python
model_data = pd.DataFrame(standard_data, columns=['followers', 'artist popularity', 'markets', 'album popularity', 'duration', 'explicit', 'popularity'])
model_data.iloc[1359]
```

```python
songs_train = model_data.iloc[50:]
songs_test = model_data.iloc[:50]

train_features = songs_train.iloc[:, :-1]
train_label = songs_train.iloc[:, -1]
test_features = songs_test.iloc[:, :-1]
test_label = songs_test.iloc[:, -1]
```
