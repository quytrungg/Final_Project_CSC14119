```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
songs_data = pd.read_csv('songs_data.tsv', sep='\t')
```

```python
songs_data.head()
```

```python
data_np = np.array(songs_data)

for i in range(len(data_np)):
    data_np[i][2] = data_np[i][2].split(',')
    data_np[i][3] = list(map(int, data_np[i][3].split(',')))
    data_np[i][4] = list(map(int, data_np[i][4].split(',')))
    data_np[i][5] = int(data_np[i][5])
    data_np[i][7] = datetime.strptime(data_np[i][7], '%Y-%m-%d')
    data_np[i][8] = int(data_np[i][8])
    data_np[i][9] = 0 if data_np[i][9] == 'False' else 1
    data_np[i][10] = 0 if data_np[i][10] == 'False' else 1
    data_np[i][11] = int(data_np[i][11])
```

```python
max = 0
idx = 0
for i in data_np:
    if len(i[2]) > max:
        max = len(i[2])
        idx = i
max, idx
```
